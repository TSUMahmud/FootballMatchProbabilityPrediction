```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
train['is_cup'] = labelencoder.fit_transform(train['is_cup'])
test['is_cup'] = labelencoder.fit_transform(test['is_cup'])
# train_y['target'].replace({"home": 0, "draw": 1, "away":2}, inplace=True)
train.drop(train.filter(regex='recorded_matches').columns, axis=1, inplace = True)
test.drop(test.filter(regex='recorded_matches').columns, axis=1, inplace = True)
```

```python
train_x = train.copy()
train_x.head()
```

```python
train_x.fillna(train_x.mean(), inplace=True)
test.fillna(test.mean(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler


sc = StandardScaler()
X_train_std = sc.fit_transform(train_x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X_train_std)
```

```python
pca_Df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
pca_Df.head()
```

```python
X_train_std
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.figure(figsize=(50,50))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis",fontsize=20)
targets = ['home', 'draw', 'away']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = train_y['target'] == target
    plt.scatter(pca_Df.loc[indicesToKeep, 'principal component 1']
               , pca_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

```python
import mglearn
fig, axes = plt.subplots(17, 2, figsize=(30, 78))
home = train_x[train_y.target == 'home']
draw = train_x[train_y.target == 'draw']
away = train_x[train_y.target == 'away']

ax = axes.ravel()

for i in range(34):
    column = train_x.columns[i]
    _, bins = np.histogram(train_x[column], bins=500)
    ax[i].hist(home[column], bins=bins, color='r', alpha=.33)
    ax[i].hist(draw[column], bins=bins, color='g', alpha=.33)
    ax[i].hist(away[column], bins=bins, color='b', alpha=.33)
    ax[i].set_title(column)
    ax[i].set_yticks(())
ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["home", "draw", 'away'], loc="best")
fig.tight_layout()
```

```python
from sklearn.manifold import TSNE


projected = TSNE(n_components=2, perplexity=50.0, learning_rate=200, random_state=123).fit_transform(X_train_std)
```

```python
import matplotlib.patheffects as PathEffects

train_y['target'].replace({"home": 0, "draw": 1, "away":2}, inplace=True)
colors = train_y['target'].copy()
f = plt.figure(figsize=(8, 8))
ax = plt.subplot(aspect='equal')
for i in range(0, 5000, 500):
    plt.scatter(projected[i:i+500, 0], projected[i:i+500, 1])

ax.axis('off')
ax.axis('tight')

txts = []

for i in range(4):
    xtext, ytext = np.median(projected[colors == i, :], axis=0)
    txt = ax.text(xtext, ytext, str(i), fontsize=24)
    txt.set_path_effects([
        PathEffects.Stroke(linewidth=5, foreground="w"),
        PathEffects.Normal()])
    txts.append(txt)
```
