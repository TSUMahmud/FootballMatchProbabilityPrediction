```python
test = pd.read_csv("/content/drive/MyDrive/7313Project/test.csv")
train = pd.read_csv("/content/drive/MyDrive/7313Project/train.csv")
testn = pd.read_csv("/content/drive/MyDrive/7313Project/test.csv")
```

```python
train.drop(train.iloc[:, [0,2,3,4,5]], axis=1, inplace=True)
```

```python
train.head()
```

```python
test.drop(test.iloc[:,0:5], axis=1, inplace=True)
```

```python
test.head()
```

```python
train.drop(train.filter(regex='date').columns, axis=1, inplace = True)
test.drop(test.filter(regex='date').columns, axis=1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
train['is_cup'] = labelencoder.fit_transform(train['is_cup'])
test['is_cup'] = labelencoder.fit_transform(test['is_cup'])
train['target'] = labelencoder.fit_transform(train['target'])
```

```python
train_y = train['target']
train_y.unique()
```

```python
train_x = train.iloc[:, 1:]
train_x.head()
```

```python
train_x.fillna(train_x.mean(), inplace=True)
test.fillna(test.mean(), inplace=True)
```

```python
train_x.describe()
```

```python
from sklearn.preprocessing import StandardScaler
#fit scaler on training data
norm = StandardScaler().fit(train_x)
#transform training data
xtrain_norm = norm.transform(train_x)
#transform testing data
test_norm = norm.transform(test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    xtrain_norm, train_y, test_size=0.3, random_state=1, stratify=train_y)

tree = DecisionTreeClassifier(criterion='gini', 
                              max_depth=6, 
                              random_state=1)
tree.fit(X_train, y_train)

y_pred = tree.predict(X_test)
acc = np.mean(y_pred == y_test)*100
acc
```

```python
from pandas.core.common import random_state
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import BaggingClassifier
import warnings
warnings.filterwarnings('ignore')

model = DecisionTreeClassifier(criterion='gini', max_depth=2)
# LinearDiscriminantAnalysis(),

param_grid = [{'n_estimators': [50, 100, 300, 500]}]

gs = GridSearchCV(estimator=BaggingClassifier(base_estimator= model, oob_score=True, bootstrap=True, bootstrap_features=False, random_state=1),
                param_grid=param_grid,
                refit=True,
                cv=3,
                n_jobs=5)

gs.fit(xtrain_norm, train_y)
print('Best Accuracy: %.2f%%' % (gs.best_score_*100))
print('Best Params:', gs.best_params_)

gs.cv_results_
```

```python
from pandas.core.common import random_state
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import BaggingClassifier
import warnings
warnings.filterwarnings('ignore')
model = DecisionTreeClassifier(criterion='gini', max_depth=1)
bc = BaggingClassifier(base_estimator=model,
                        n_estimators=100,
                        oob_score=True,
                        bootstrap=True,
                        bootstrap_features=False,
                        n_jobs=10,
                        random_state=1)
bc.fit(xtrain_norm, train_y)
y_pred = bc.predict(xtrain_norm)
train_acc = np.mean(y_pred == train_y)*100
train_acc
```

```python
from mlxtend.evaluate import confusion_matrix
confmat = confusion_matrix(train_y, y_pred)

print(confmat)
```

```python
prediction_1 = bc.predict_proba(test_norm)
prediction_1
```

```python
submission = pd.concat([pd.DataFrame(testn['id']), pd.DataFrame(prediction_1, columns=['home', 'draw', 'away'])[['home', 'draw', 'away']]], axis=1)
submission.to_csv('/content/drive/MyDrive/7313Project/submission_bagging.csv', index=False)
```
